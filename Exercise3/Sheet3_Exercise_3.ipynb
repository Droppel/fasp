```python
import matplotlib.pyplot as plt
import numpy as np
import math

def roots_of_unity(n):
    """
    Calculates the nth roots of unity.

    Parameters:
    n (int): The number of roots.

    Returns:
    list of complex: The nth roots of unity.
    """
    roots = [np.exp(2j * np.pi * k / n) for k in range(n)]
    regulated=[]
    for r in roots:
      if abs(r.real)<1e-10:
        r=complex(0,r.imag)
      if abs(r.imag)<1e-10:
        r=complex(r.real,0)
      regulated.append(r)

    return regulated

def primitive_roots_of_unity(n):
    """
    Calculates the nth primitive roots of unity.

    Parameters:
    n (int): The number of roots.

    Returns:
    list of complex: The nth primitive roots of unity.
    """
    primitive_roots = []
    for r in roots_of_unity(n):
        if is_primitive_root(r, n):
            primitive_roots.append(r)
    return primitive_roots

def is_primitive_root(r, n):
    """
    Returns:
    bool: True if r is a primitive nth root of unity, False otherwise.
    """

    for k in range(1,n):
      if pow(r,k) == 1:
        return False
    return True

def plot_roots_of_unity(n):
    """
    Plots the nth roots of unity with different colors for primitive and non-primitive roots.

    Parameters:
    n (int): The number of roots.
    """
    roots = roots_of_unity(n)
    primitive_roots = primitive_roots_of_unity(n)

    x = [r.real for r in roots]
    y = [r.imag for r in roots]
    plt.figure(figsize=(4, 4))
    plt.plot(x, y, 'o', color='blue')
    plt.plot([r.real for r in primitive_roots], [r.imag for r in primitive_roots], 'o', color='red')

    plt.xlabel('Re')
    plt.ylabel('Im')

    circle = plt.Circle((0,0), 1,color='k',edgecolor='b',alpha=0.2)



    ax = plt.gca()
    ax.set_ylim([-1.2, 1.2])
    ax.set_xlim([-1.2, 1.2])
    ax.add_artist(circle)
    ax.grid(linestyle = "dashed")
    plt.title(f'Roots of unity, n = {n}')
    plt.legend(['Non-primitive roots', 'Primitive roots'])
    plt.show()
# Plot the roots of unity for n = 1 to 10
for n in range(1,10):
    plot_roots_of_unity(n)
```
